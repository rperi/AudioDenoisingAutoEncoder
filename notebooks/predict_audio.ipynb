```python
import os
from keras.models import load_model
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
model = load_model('../models/RawAudioCDNNAutoEncoder_7/weights-improvement-29--0.00.hdf5')
```

```python
import soundfile as sf
import numpy as np
from IPython.display import Audio
N = 16384
```

```python
sound, sr = sf.read('../data/audio/clean/sp02.wav')
sound = sound[:N].astype('float32')
Audio(sound, rate=sr)
```

```python
sound.min()
```

```python
s = sound.reshape(1,N,1)
p = model.predict(s)
```

```python
plt.plot(sound[100:15000])
```

```python
p = p.reshape(N, 1)
plt.plot(p[100:15000]*2000 - 0.25)
```
